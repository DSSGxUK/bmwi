```python
import pandas as pd
import geopandas as gpd
```

```python
df_index = pd.read_csv('../data/index.csv')
```

```python
def fix_ags5(x):
    if len(str(x))==4:
        return '0'+str(x)
    else:
        return str(x)
```

```python
df_index['ags5_fix'] = df_index['ags5'].apply(fix_ags5)
```

```python
gdf = gpd.read_file('../georef-germany-kreis/georef-germany-kreis-millesime.shp')
```

```python
# fix gdf krs_name to our kreis name
gdf['krs_name']=df_index['kreis']
```
