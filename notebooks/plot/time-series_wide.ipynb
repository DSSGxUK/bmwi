```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../../data/df_final_date_wide_2007.csv', index_col=0)
df.shape
```

```python
list(df.columns)
```

```python
filter_col = 'ags5'
filter_val = 1001
filter_df = df[df[filter_col]==filter_val]
```

```python
x_col = 'date'
y_col = 'unemployment_rate' 
```

```python
plt.figure(figsize=(50,20))
plt.plot(filter_df[x_col], filter_df[y_col], label=f'{filter_col}={filter_val}')
plt.xlabel(f'{x_col}')
plt.ylabel(f'{y_col}')
plt.xticks(rotation=60)
plt.legend()
```

```python
def plot_line1(df, x_col, y_col, fliter_col=None, filter_val=None):
    plt.figure(figsize=(50,20))
    if (filter_col==None) or (filter_val==None):
        plt.plot(df[x_col], df[y_col])
    else:
        filter_df = df[df[filter_col]==filter_val]
        plt.plot(filter_df[x_col], filter_df[y_col], label=f'{filter_col}={filter_val}')
    plt.xlabel(f'{x_col}')
    plt.ylabel(f'{y_col}')
    plt.xticks(rotation=60)
    plt.legend()
    plt.show()
```

```python
def plot_line(df, x_col, y_col, fliter_col=None, filter_val=None):
    fig, ax = plt.subplots(figsize=(50,20))
    if (filter_col==None) or (filter_val==None):
        df.plot(x_col, y_col, ax=ax)
        ax.set_title(f'{y_col} in Germany')
    else:
        filter_df = df[df[filter_col]==filter_val]
        filter_df.plot(x_col, y_col, ax=ax)
        ax.set_title(f"{filter_val}'s {y_col} in Germany")
    return fig
```

```python
plot_line1(df, 'date', 'unemployment_rate')
```

```python
plot_line(df, 'date', 'unemployment_rate')
```

```python
plot_line1(df, 'date', 'unemployment_rate', 'ags5', 1001)
```

```python
plot_line(df, 'date', 'unemployment_rate', 'ags5', 1001)
```
