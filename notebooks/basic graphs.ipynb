```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
bevo = pd.read_csv('bevoelkerung.csv')
bevo.columns
```

# Histogram

```python
def data_histogram(column, no_of_bins, minimum=None, maximum=None):
    
    # default vals
    if minimum is None:
        minimum = min(column)
    if maximum is None:
        maximum = max(column)
        
    # calculate the histogram edges
    bin_edges = np.arange(minimum, maximum, (maximum - minimum)/no_of_bins)
    
    print(bin_edges)
    
    plt.hist(column, bin_edges)
      
    plt.show()
```

```python
data_histogram(bevo['kr_ew_19'], 10, None, 1000000)
```

# lmplot

```python
def lmplot(db, col1, col2):
    sns.lmplot(col1, col2, data=db, fit_reg=False)
```

```python
lmplot(bevo, 'kr_ew_w', 'kr_ew_m')
```

```python
def heatmap(db, col_names_list=None):
        
    if col_names_list is None:
        col_names_list = db.columns
    
    f, ax = plt.subplots(figsize=(11,11))
    sns.heatmap(db[col_names_list].corr(), annot=True)
```

```python
col_names_list = ['_id', 'ags2', 'bundesland', 'ags5', 'kreis', 'kr_geb', 'kr_ew_18',
       'kr_ew_19', 'kr_ew_m', 'kr_ew_w', 'kr_ew_00u03', 'kr_ew_03u06',
       'kr_ew_06u10', 'kr_ew_10u15']
heatmap(bevo, col_names_list)
```
